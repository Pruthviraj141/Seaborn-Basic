```python
# titanic_analysis.py or notebook cell

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# (Optional) set style
sns.set_style("whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)

# Load data
df = pd.read_csv('train.csv')  # change path if needed

# Quick look
print(df.head())
print(df.info())
print(df.describe())

# Check missing values
print(df.isnull().sum())

# 1. Count plot: how many survived vs not survived
sns.countplot(x='Survived', data=df)
plt.title('Survival Count on Titanic')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()

# 2. Survival by gender
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Survival by Gender')
plt.show()

# 3. Age distribution
sns.histplot(df['Age'].dropna(), bins=30, kde=True)
plt.title('Age Distribution of Passengers')
plt.show()

# 4. Boxplot: Age by passenger class
sns.boxplot(x='Pclass', y='Age', data=df)
plt.title('Age Distribution by Passenger Class')
plt.show()

# 5. Violin plot: Fare by class & survival
sns.violinplot(x='Pclass', y='Fare', hue='Survived', data=df, split=True)
plt.title('Fare Distribution by Class and Survival')
plt.ylim(0, 300)  # restrict Y for better visibility
plt.show()

# 6. Correlation heatmap for numeric features
numeric_cols = ['Survived','Pclass','Age','SibSp','Parch','Fare']
corr = df[numeric_cols].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation between numeric features')
plt.show()
```
